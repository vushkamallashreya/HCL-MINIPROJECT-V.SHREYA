LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df_train=df_train.copy()
```

```python
df_train['is_promoted'].value_counts()
```

```python
df_train['get_promoted'] = df_train.is_promoted.map(lambda x: 1 if x == 1 else 0)
df_train['not_get_promoted'] = df_train.is_promoted.map(lambda x: 1 if x == 0 else 0)
```

```python
features_list = ['KPIs_met >80%', 'awards_won?', 'previous_year_rating',
                 'get_promoted', 'not_get_promoted'] 
X_train, X_test, y_train, y_test = train_test_split(df_train[features_list],
                                                      df_train.avg_training_score, 
                                                        test_size=0.20, 
                                                        random_state=997)
model = LinearRegression()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(X_train)

# Impute the data, then train the data
X_train_imp = imp.transform(X_train)
X_test_imp = imp.transform(X_test)

model = model.fit(X_train_imp, y_train)
predict_array = model.predict(X_test_imp)
```

```python
print(mean_absolute_error(y_test, predict_array))
print(mean_squared_error(y_test, predict_array))
print(r2_score(y_test, predict_array))
```

```python
df_train['department'].value_counts()
```

```python
departments = ['Sales & Marketing', 'Operations', 'Technology', 'Procurement', 'Analytics', 'Finance', 'HR', 'Legal', 'R&D']
for dep in departments: 
    # use dep to avoid being confused with department column
    df_train[dep] =df_train.department.map(lambda x: 1 if x == dep else 0)
```

```python
 df_train['region'].value_counts()
```

```python
regions = []
for i in range(1, 35):
    regions.append('region_' + str(i))

# Use for loop to add 34 columns, one for each region
for reg in regions:
    df_train[reg] =  df_train.region.map(lambda x: 1 if x == reg else 0)
```

```python
 df_train['education'].value_counts()
```

```python
educations = ["Bachelor's ", "Master's & above", "Below Secondary"]
for edu in educations:
    df_train[edu] = df_train.education.map(lambda x: 1 if x == edu else 0)
```

```python
df_train['gender'].value_counts()
```

```python
genders = ['f', 'm']
for gen in genders:
    df_train[gen] = df_train.gender.map(lambda x: 1 if x == gen else 0)
```

```python
df_train['recruitment_channel'].value_counts()
```

```python
channels = ['other', 'sourcing', 'referred']
for cha in channels:
    df_train[cha] = df_train.recruitment_channel.map(lambda x: 1 if x == cha else 0)
```

```python
df_train.head()
```

```python
features_list = ['KPIs_met >80%', 'get_promoted', 'not_get_promoted', 
                 'awards_won?', 'previous_year_rating', 'no_of_trainings', 'age',
                 'length_of_service'] + departments + regions + genders + channels + educations
X_train1, X_test1, y_train1, y_test1 = train_test_split(df_train[features_list],
                                                        df_train.avg_training_score, 
                                                        test_size=0.20, 
                                                        random_state=997)
model1 = LinearRegression()

# Create imputer to replace missing values with the mean 
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(X_train1)

# Impute the data, then train the data
X_train_imp1 = imp.transform(X_train1)
X_test_imp1 = imp.transform(X_test1)

model1 = model1.fit(X_train_imp1, y_train1)
predict_array1 = model1.predict(X_test_imp1)
```

```python
print(mean_absolute_error(y_test1, predict_array1))
print(mean_squared_error(y_test1, predict_array1))
print(r2_score(y_test1, predict_array1))
```

```python
def predict_promotion(no_of_trainings,age,previous_year_rating,length_of_service,avg_training_score,department,education,recruitment_channel):    
    loc_index = np.where(X.columns==department)[0][0]
    loc_index1 = np.where(X.columns==education)[0][0]
    loc_index2 = np.where(X.columns==recruitment_channel)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = no_of_trainings
    x[1] = age
    x[2] = previous_year_rating
    x[3]= length_of_service
    x[4] = avg_training_score
    if loc_index >= 0:
        x[loc_index] = 1
    if loc_index1 >= 0:
        x[loc_index1] = 1
    if loc_index2 >= 0:
        x[loc_index2] = 1

    return rfc.predict([x])[0]
```
